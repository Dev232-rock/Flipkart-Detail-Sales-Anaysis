```python
top_products = pd.DataFrame(a['product_name'].value_counts()[:10]).reset_index() #these are top 10 products based on count
top_products
```

```python
top_brands = pd.DataFrame(a['brand_name'].value_counts() [:10]).reset_index()  #these are the top 10 brand names
top_brands
```

```python
# @title count

from matplotlib import pyplot as plt
top_brands['count'].plot(kind='hist', bins=20, title='count')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
a_brand = a[a['brand_name'] == 'Flipkart'].reset_index() #there are 9 products under flipkart brand
a_brand
```

```python
a.groupby('l0_category').size().plot( kind = 'pie' , autopct ='%.1f' , figsize = (12,10)) #this includes the l0_category list
```

```python
brand_counts = a['brand_name'].value_counts()# Top 20 brands which has higher presence in market with high number of products
top_20_brands = brand_counts.nlargest(20)

colors = ['skyblue', 'orange', 'green', 'purple', 'pink', 'yellow', 'brown', 'teal', 'gray', 'magenta',
          'lime', 'cyan', 'blue', 'red', 'gold', 'indigo', 'maroon', 'turquoise', 'coral', 'lightgreen']
plt.figure(figsize=(12, 6))
plt.bar(top_20_brands.index, top_20_brands.values, color=colors)
plt.xlabel('Brand')
plt.ylabel('Number of products') #no of products is the product sold
plt.title('Brand Presence in the Market')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))   #here are the top product names based on count
plt.bar(top_products['product_name'],top_products['count'], color = "green" , label='Count')
plt.title('Top 10 product names ')
plt.xlabel('product_name')
plt.ylabel('count')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30,18))
l2_category =a.l2_category.value_counts()[0:10]  #top 1
plt.bar(l2_category.index,l2_category)
plt.xlabel = ('Category')
plt.title  = ('Top 10 categories')
plt.show()

```
